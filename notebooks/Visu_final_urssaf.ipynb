"""
# Calcul du Score Sectoriel
Ce script calcule un score sectoriel basé sur plusieurs paramètres, permettant d'évaluer le dynamisme économique d'un secteur.

## Composantes du score :
1. **Part des effectifs** : Part des effectifs du secteur par rapport au total des effectifs.
2. **Part des établissements** : Part des établissements du secteur par rapport au total des établissements.
3. **Taux de croissance** : Variation relative des effectifs et établissements d'une année sur l'autre.
4. **Indice de localisation** : Mesure de la concentration d'un secteur dans une région par rapport à la moyenne nationale.
5. **Taux de nouvelles entreprises** : Proportion des entreprises créées par rapport au total d'entreprises dans le secteur.

## Formule du score sectoriel :
Le score est une moyenne pondérée des différentes composantes :


```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Calcul de la part des effectifs et des établissements
df['part_effectifs'] = df['nb_effectif'] / df['nb_effectif_total']
df['part_etablissements'] = df['nb_entreprise'] / df['nb_entreprise_total']

# Taux de croissance des effectifs et établissements (par secteur et ville)
df['croissance_effectifs'] = df.groupby(['ville', 'secteur_na88'])['nb_effectif'].pct_change()
df['croissance_etablissements'] = df.groupby(['ville', 'secteur_na88'])['nb_entreprise'].pct_change()

# Indice de localisation (par secteur et région)
# Moyenne nationale des parts pour normaliser
moyenne_nationale_effectifs = df.groupby('secteur_na88')['part_effectifs'].transform('mean')
df['indice_localisation'] = df['part_effectifs'] / moyenne_nationale_effectifs

# Taux de nouvelles entreprises
df['taux_nouvelles_entreprises'] = df.groupby(['ville', 'secteur_na88'])['nb_entreprise'].diff() / df.groupby(['ville', 'secteur_na88'])['nb_entreprise'].shift(1)

# Remplacer les valeurs infinies ou NaN résultant des divisions par zéro ou absences de données
df.fillna(0, inplace=True)
df.replace([np.inf, -np.inf], 0, inplace=True)

# Poids des composantes
poids_effectifs = 0.3
poids_etablissements = 0.3
poids_croissance = 0.2
poids_localisation = 0.1
poids_nouvelles_entreprises = 0.1

# Calcul du score sectoriel
df['score_sectoriel'] = (
    (poids_effectifs * df['part_effectifs']) +
    (poids_etablissements * df['part_etablissements']) +
    (poids_croissance * (df['croissance_effectifs'] + df['croissance_etablissements']) / 2) +
    (poids_localisation * df['indice_localisation']) +
    (poids_nouvelles_entreprises * df['taux_nouvelles_entreprises'])
)

# Agrégation pour la visualisation : moyenne des scores par secteur et année
df_scores = df.groupby(['secteur_na88', 'année'])['score_sectoriel'].mean().reset_index()

# Visualisation de l'évolution des scores sectoriels
plt.figure(figsize=(14, 8))
sns.lineplot(data=df_scores, x='année', y='score_sectoriel', hue='secteur_na88', marker='o')
plt.title('Évolution des scores sectoriels par année', fontsize=16)
plt.xlabel('Année', fontsize=14)
plt.ylabel('Score sectoriel', fontsize=14)
plt.legend(title='Secteurs', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
