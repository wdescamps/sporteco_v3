```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('/content/Feuille de calcul sans titre - Feuille 1 (4).csv')
```

```python
df
```

```python
grouped_df = df.groupby('annee', as_index=False).agg({
    'score_sportif_norm': 'mean',
    'score_eco_norm': 'mean',})
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=grouped_df['annee'],
        y=grouped_df['score_sportif_norm'],
        mode='lines+markers',
        name='Score Sportif',
        line=dict(color='blue'),
        hovertemplate='Saison: %{x}<br>Score Sportif Moyen: %{y:.2f}<br>Classement Moyen: %{customdata}'
    )
)

fig.add_trace(
    go.Scatter(
        x=grouped_df['annee'],
        y=grouped_df['score_eco_norm'],
        mode='lines+markers',
        name='Score Économique',
        line=dict(color='green'),
        hovertemplate='Saison: %{x}<br>Score Économique: %{y:.2f}'
    )
)

fig.update_layout(
    title="Évolution Moyenne du Score Sportif",
    xaxis_title="Saison",
    yaxis_title="Score Sportif Moyen",
    hovermode="x unified"
)

fig.show()
```

```python
df=df.sort_values(by= 'annee')

# Liste des villes uniques
villes = df['ville'].unique()

# Créer la figure
fig = go.Figure()

# Ajouter les traces par défaut (première ville)
ville_initiale = villes[0]
df_ville = df[df['ville'] == ville_initiale]

fig.add_trace(
    go.Scatter(
        x=df_ville['annee'],
        y=df_ville['score_sportif_norm'],
        mode='lines+markers',
        name='Score Sportif',
        line=dict(color='blue'),
        hovertemplate='Saison: %{x}<br>Score Sportif Moyen: %{y:.2f}'
    )
)

fig.add_trace(
    go.Scatter(
        x=df_ville['annee'],
        y=df_ville['score_eco_norm'],
        mode='lines+markers',
        name='Score Économique',
        line=dict(color='green'),
        hovertemplate='Saison: %{x}<br>Score Économique: %{y:.2f}'
    )
)

# Ajouter les boutons pour le dropdown
dropdown_buttons = []
for ville in villes:
    dropdown_buttons.append(
        dict(
            args=[
                {"x": [df[df['ville'] == ville]['annee'], df[df['ville'] == ville]['annee']],
                 "y": [df[df['ville'] == ville]['score_sportif_norm'], df[df['ville'] == ville]['score_eco_norm']],
                 "name": [f"{ville} - Score Sportif", f"{ville} - Score Économique"]}],
            label=ville,
            method="update"
        )
    )

# Mettre à jour le layout avec le menu
fig.update_layout(
    updatemenus=[
        dict(
            buttons=dropdown_buttons,
            direction="down",
            showactive=True,
            x=0.5,
            xanchor="center",
            y=1.2,
            yanchor="top"
        )
    ],
    title="Évolution Moyenne des Scores Sportif et Économique par Ville",
    xaxis_title="Saison",
    yaxis_title="Score Normalisé",
    hovermode="x unified"
)

fig.show()
```
